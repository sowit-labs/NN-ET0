# Imports

```python
import pandas as pd
from NN_ET0.NN_ET0_helpers import get_openweather_past5days, get_openweather_7daysforecast, modelName_from_coordinates, train_ANN, perform_ET0_prediction
from pcse.db import NASAPowerWeatherDataProvider
import matplotlib.pyplot as plt
```

# Entry variables

```python
## entering coordinates for Marrakech
latitude = -8.001
longitude = 31.636 # 31.636

## weather forecast API key
# free tier for 7-day forecast
# API key can be obtained for free upon registration at https://openweathermap.org
api_key = ""

## ANN parameters
# Model class selection
# For data provided by OpenWeatherMap, models M5, M6, M8 can be trained but not be used 
# for prediction, as IRRAD variable is not available in the forecast.
# M1 : TEMP, TMIN, TMAX, RAIN
# M2 : TEMP, TMIN, TMAX, RAIN, VAP
# M3 : TEMP, TMIN, TMAX, RAIN, WIND
# M4 : TEMP, TMIN, TMAX, RAIN, VAP, WIND
# M5 : TEMP, TMIN, TMAX, RAIN, VAP, WIND, IRRAD
# M6 : TEMP, TMIN, TMAX, IRRAD
# M7 : TEMP, TMIN, TMAX, WIND
# M8 : TEMP, TMIN, TMAX, RAIN, IRRAD
model_class = "M4" 

# model name, storage path
path='../data/'
```

# Gathering weather info on which to perform ET0 estimation

## Past 5 days

```python
df_weather_past = get_openweather_past5days(latitude, longitude, api_key)
df_weather_past
```

## 7-day forecast

```python
df_weather_forecast = get_openweather_7daysforecast(latitude, longitude, api_key)
df_weather_forecast
```

## Combining datasets

```python
prediction_df = df_weather_past.append(df_weather_forecast).reset_index(drop=True)
prediction_df
```

# Performing training & prediction

```python
dfPred = perform_ET0_prediction(model_class, latitude, longitude, prediction_df, loops=5)
```

```python
dfPred
```

# Graphing output

```python
# on récupère les données NASA Power
weatherdata = NASAPowerWeatherDataProvider(longitude,latitude) 
print(weatherdata)

df = pd.DataFrame(weatherdata.export())
# converting PCSE units : cm to mm
df["ET0"]=df["ET0"]*10 
df_all = df[-60:].rename(columns={"DAY": "TIMESTAMP_DAY"}).append(dfPred)
```

```python
plt.figure(num=None, figsize=(16, 8), dpi=150, facecolor='w', edgecolor='k')
plt.plot(df_all["TIMESTAMP_DAY"],df_all["ET0"],linewidth=.5)
plt.plot(df[-60:]["DAY"],df[-60:]["ET0"], label="NASA Power ET0 regional values, up to -90 days in the past")
plt.plot(df_all.loc[df_all["type"]=="past","TIMESTAMP_DAY"],df_all.loc[df_all["type"]=="past","ET0"],label="average ET0 estimation ("+model_class+" model) from local weather forecast, past 5 days",linewidth=2)
plt.plot(df_all.loc[df_all["type"]=="forecast","TIMESTAMP_DAY"],df_all.loc[df_all["type"]=="forecast","ET0"],label="average ET0 estimation ("+model_class+" model) from local weather forecast, 7 days forecast",linewidth=2)

plt.xticks(df_all["TIMESTAMP_DAY"], rotation=70)
plt.title("ET0 for coordinates lat,long "+str(latitude)+","+str(longitude))
plt.legend() 
plt.show()
```
